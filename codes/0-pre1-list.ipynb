# Practice Exercises

1. Write the first c++ program to understand the structure of the program.
2. Write a program to draw a given material property table by using cout function and escape sequences.
3. Write a program to draw a material property table for four materials using values entered by the user.
4. Write a program to draw a given material property table using functions and values entered by the user.
5. Write a program to calculate the sum of two forces using the function with argument approach implementing the calling of function with call by value type.
6. Write a program to find distance between two points from its coordinates.
7. Write a program for finding the magnitude of forces by Lami’s theorem.
8. Write a program to resolve the inclined force using functions.
9. Write a program to find the resultant of two forces normal to each other using functions.
10. Write a program to find the resultant of two coplanar and concurrent forces using functions.
11. Write a program to calculate the roots of a quadratic equation using decision control statements in perspective to work with all the cases whether the equation gives a real, imaginary or equal roots.
12. Write a program to find the linear speed of a conveyor belt with provision for conversion of units as per user input using decision control statements and operators.
13. Write a program to categorise the regime as per the calculated Mach Number.
14. Write a program to find the result of placing a solid body into a liquid with the flexibility to adapt to any liquid and solid.
15. Write a program for metric conversions of various properties: temperature, pressure, length, energy, and power using a switch statement.
16. Write a program to draw a given material property table by using iterative statements and arrays.
17. Write a program to find the properties of a material from the tensile test by providing the load, elongation, and geometrical properties of a specimen by using functions, arrays, and iterative statements.
18. Write a program to find the properties of a material from the tensile test by providing the load, elongation, and geometrical properties of a specimen by using functions, arrays, and iterative statements with the choice of getting input from sample sets of readings obtained from a text (.txt) file or user entered values.
19. Write a program to calculate the area of a rectangle and a circle using pointers.
20. Write a program to calculate the area of a rectangle and a circle using call by reference, and structures.
