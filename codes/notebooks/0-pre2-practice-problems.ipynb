# Practice Problems

1. Practice Assignment 1(1): Write a c++ program to print a physical law.
2. Practice Assignment 2(2): Write a program to draw a thermal conductivity table for two different materials by using cout function and escape sequences.
3. Practice Assignment 3(3): Write a program to draw a thermal conductivity table for two different materials and values should be entered by the user.
4. Practice Assignment 4(4): Write a program to draw a thermal conductivity table for two different materials and values should be entered by the user. The program should be written using functions.
5. Practice Assignment 5(5): Write a program to calculate the sum of two velocities using function with argument approach. Calling of function will be of call by value type.
6. Practice Assignment 6(6-10): Write a program to compute the resultant of two forces using the parallelogram law of vector addition.
7. Practice Assignment 7(11-15): Write a program for metric conversions of temperature and pressure using if else statement for chossing between the two and switch statement for choosing between the units.
8. Practice Assignment 8(16-17): Write a program to draw a given thermal conductivity table by using iterative statements, arrays, and array of strings.
9. Practice Assignment 9(18): Write a program to draw a given thermal conductivity table by using iterative statements, arrays, and array of strings. The sample sets of readings are obtained from text (.txt) file; also give the option if the user wants to enter the values.
